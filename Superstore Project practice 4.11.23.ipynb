```python
# Plot a histogram
sns.histplot(df['Ship Mode'], kde=True)
plt.show()
```

```python
plt.hist(df['Postal Code'], bins=10, color='blue', alpha=0.7)
plt.title('Histogram of Postal codes')
plt.xlabel('NumericColumn values')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='Sub-Category', data=df)
plt.title('Bar Chart of Subcategory')
plt.show()
```

```python
sns.countplot(x='Quantity', data=df)
plt.title('Bar Chart of Quantity')
plt.show()
```

```python
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Scatter Plot of Sales and Profit')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Segment'].value_counts(), labels=df['Segment'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Segment')
plt.show()
```

```python
sns.boxplot(x='Quantity', y='Category', data=df)
plt.title('Box Plot of Quantity by Category')
plt.show()
```

```python
import seaborn as sns
sns.violinplot(x='Category', y='Sales', data=df) #violin plot Combines aspects of box plot and kernel density plot, providing a richer description of the distribution.
```

```python
import seaborn as sns
sns.heatmap(df.corr(), annot=True, cmap='coolwarm') #heatmap Useful for displaying relationships or correlations between variables in a matrix format.
```

```python
sns.pairplot(df, hue='Category') #pairplot Displays pairwise relationships in a dataset, great for exploring correlations.
```

```python
ax = plt.subplot(111, polar=True)
ax.plot(df['Order Date'], df['Sales'])
ax.set_xticks(np.arange(0, 2*np.pi, np.pi/6))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show() #polar plot Suitable for displaying cyclic data, like time-series data.
```

```python
plt.errorbar(df['Category'], df['Sales'], yerr=df['Sales'].std(), fmt='o', capsize=5)
plt.show() #error bar Shows the variability of data and can be used to indicate uncertainty.
```

```python
plt.pie(df['Category'].value_counts(), labels=df['Category'].unique(), autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.axis('equal')
plt.show() #donut chart Similar to a pie chart but with a hole in the center. It can be used to show the proportion of each category in a dataset.

```

```python
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)
X, Y = np.meshgrid(x, y)
U = np.sin(X)
V = np.cos(Y)

plt.streamplot(X, Y, U, V, density=1, linewidth=1, color=U, cmap='viridis')
plt.show()
#streamplot Useful for visualizing 2D vector fields, showing flow and direction.
```

```python
import plotly.express as px

# ... (prepare hierarchical data)
# ...

fig = px.sunburst(df, path=['Sub-Category', 'Segment', 'Category'], values='Sales')
fig.show()
#sunburst chart Ideal for displaying hierarchical data in a radial layout.
```

```python
sns.kdeplot(df['Sales'], shade=True)
plt.show() #KDE(kernel density estimation) plot Visualizes the probability density function of a continuous random variable.
```

```python
plt.fill_between(df['Order Date'], df['Sales'], color='skyblue', alpha=0.4)
plt.plot(df['Order Date'], df['Sales'], color='skyblue', alpha=0.8)
plt.show()
#area chart Shows the trend of a numerical variable over a range. Useful for visualizing cumulative quantities.
```

```python
plt.hexbin(df['Discount'], df['Quantity'], gridsize=20, cmap='Blues')
plt.colorbar()
plt.show()
#hexbin plot Useful for visualizing the distribution of a large number of points in a 2D space.
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Sales'], df['Profit'], df['Quantity'])
plt.show() #3d scatter plot Adds an extra dimension to a scatter plot, useful when you have three numerical variables.
```
